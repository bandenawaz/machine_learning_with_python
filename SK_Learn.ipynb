# **Data**

The Scikit learn comes with some standard machine learning datasets.

**load dataset into the memory**

```python
from sklearn.datasets import load_wine

wine_data = load_wine()
```

```python
#lets access the data
wine_data.data
```

```python
#lets convert the array into pandas dataframe
import pandas as pd
wine_df = pd.DataFrame(wine_data.data, columns=wine_data.feature_names)

#Add the Target Label
wine_df["target"] = wine_data.target

#Lets view the data
wine_df.head()
```

```python
#print last 20 rows of the dataframe
wine_df.tail(20)
```

```python
#print the no of rows and columns of the dataframe
wine_df.shape
```

```python
#print the description of the dataframe
wine_df.describe
```

# Data Exploration

Pandas Dataframe is defined as 2D labeled data structure consisting of rows and columns.

```python
#print the information abt the dataframe
wine_df.info()
```

# Data Preprocessing

The data from the real world is messy, as it may contain:


*   Missing Values
*   Redundant Values
*   Outliers
*   Errors
*   Noise



```python
#Lets standardize the features using sklearn's StandardScaler
from sklearn.preprocessing import StandardScaler

#lets split the data into features and labels
features = wine_df[wine_data.feature_names].copy()
labels = wine_df['target'].copy()

#lets instantiate the scaler
scaler = StandardScaler()

#lets train our model
scaler.fit(features)

#lets transform the features
scaled_features = scaler.transform(features.values)

#lets view the first instance of transformed features
print(scaled_features[0])
```

# Model Training


```python
#lets split the data into training and test set
from sklearn.model_selection import train_test_split

X_train_scaled, X_test_scaled, y_train, y_test = train_test_split(scaled_features, labels, train_size=.8,random_state=25)

#lets checkout the splits are correct or not
print(f"Train Size : {round(len(X_train_scaled) /len(features) * 100)}% \n\
Test Size: {round(len(X_test_scaled)/len(features)*100)}%")
```

```python
print("Train Size: {0}% \nTest Size :{1}%".format(round(len(X_train_scaled) /len(features) * 100),
                                                    round(len(X_test_scaled)/len(features)*100)))
```

# Building the Model

These are the algos which we will be using

* Logistic Regression
* SVM
* Decision Tree Classifier

```python
#train using logistic regression
from sklearn.linear_model import LogisticRegression
#svm
from sklearn.svm import SVC
#Decison Tree Classifier
from sklearn.tree import DecisionTreeClassifier

#Instantiating the model
log_reg = LogisticRegression()
#svm
svm = SVC()
#Decison Tree Classifier
dtc = DecisionTreeClassifier()

#Train the model
log_reg.fit(X_train_scaled, y_train)
#svm
svm.fit(X_train_scaled, y_train)
#Decison Tree Classifier
dtc.fit(X_train_scaled, y_train)

#predict the trained model
log_reg_pred = log_reg.predict(X_test_scaled)
#svm
svm_pred = svm.predict(X_test_scaled)
#Decison Tree Classifier
dtc_pred = dtc.predict(X_test_scaled)
```

```python
print("Logistic Regression ",log_reg_pred)
print("SVM :",svm_pred)
print("Decision Tree Classifier ",dtc_pred)
```

# Model Evaluation

```python
from sklearn.metrics import classification_report

#Store model predictions in a dicitionary
#this makes it easier ot iterate through each model
#and print the results
model_preds = {
    'Logistic Regression': log_reg_pred,
    'Support Vector Machine': svm_pred,
    'Decision Tree Classifier': dtc_pred
}

for model, preds in model_preds.items():
  print("{0} Results: \n{1}".format(model, classification_report(y_test, preds)),sep="\n\n")
```
